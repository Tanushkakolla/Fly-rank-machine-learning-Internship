## 1. My lane (or freestyle) and why

```python
# My Lane

**Selected Lane:** Refresh / Content Opportunity Scoring

I selected the **Refresh / Content Opportunity Scoring** lane because it focuses on a practical business decision: identifying which webpages should be reviewed first for content refresh. Instead of checking every page manually, a machine learning model can rank pages based on their likelihood of needing attention. The goal is to support SEO and content teams in prioritizing their work more efficiently and making better data-driven decisions.
```

## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

```python
## Decision

The decision is determining which webpages should be prioritized for content refresh instead of reviewing every page manually.

## Action

The SEO or content team can use the ranked list to update the highest-priority pages first, helping them use their time and resources more effectively.

## Cost of a wrong recommendation

If the model recommends the wrong pages, valuable time may be spent updating content that does not need immediate attention. If it misses pages that are actually declining, those pages may continue losing traffic and search visibility.

## Why ML helps

Many factors such as impressions, CTR, search position, content age, and word count interact together. Machine learning can analyze these signals simultaneously and provide better ranking recommendations than simple manual rules.
```

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

print("Total Pages:", len(df))
print("Average CTR:", round(df["ctr"].mean(), 3))
print("Median Word Count:", int(df["word_count"].median()))

declining_pct = (df["trend_direction"] == "down").mean() * 100
print(f"Declining Pages: {declining_pct:.1f}%")

corr = df["search_volume"].corr(df["impressions_90d"])
print(f"Search Volume vs Impressions Correlation: {corr:.3f}")
```
